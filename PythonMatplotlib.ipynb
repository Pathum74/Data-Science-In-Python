```python
import matplotlib.pyplot as plt
import numpy as nm
import pandas as pd
```

```python
data=pd.read_csv('srilanka covid data.csv')
data.describe()
```

```python
data.dtypes
```

Two ways to create matplotlib plot
1.Funtional approach - use command

```python
plt.plot(data['time line'],data['total_cases'])
```

2.Object orianted interface - creating a figure object and with in figure 

```python
fig=plt.figure()
ax1=fig.add_axes([0.5,0.2,1,1])

ax1.plot(data['time line'],data['total_cases'])
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,3,1])
ax.plot(data['time line'],data['total_cases'])
```

```python
fig,ax=plt.subplots()
ax.plot(data['time line'],data['total_cases'],label='total cases')
ax.set_xlabel('time line')
ax.set_ylabel('total cases')
ax.set_title('Covid Cumalative')
ax.legend(loc='best')
```
